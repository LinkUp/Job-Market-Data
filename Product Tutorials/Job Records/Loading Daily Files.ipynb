# Loading Daily Files

##### Purpose:  To provide code for loading the daily job descriptions files that are published daily and compressed in a tarball

Once the file is downloaded to disk, we will use python for decompression in memory.  We will load it into a pandas dataframe.

```python
# Libary imports
import tarfile
import pandas as pd
```

```python
# Open tarfile
with tarfile.open('raw_daily_2019-09-23.tar.gz', "r:*") as tar:
    # Get path to the job_records file within the tarfile
    csv_path = tar.getnames()[0]
    # Load job records file into pandas dataframe
    df = pd.read_csv(tar.extractfile(csv_path), escapechar='\\')

# View Results
df.head(10)
```
